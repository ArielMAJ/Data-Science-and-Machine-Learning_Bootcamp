```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('cost_revenue_clean.csv')
```

```python
data.describe()
```

```python
X = pd.DataFrame(data, columns=['production_budget_usd'])
y = pd.DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
(3e9==3000000000, 4.5e8==450000000)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3e9)
plt.xlim(0, 4.5e8)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(X, y)
```

Slope coefficient:

```python
regression.coef_    # theta_1
```

```python
#Intercept
regression.intercept_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)

# Adding the regression line here:
plt.plot(X, regression.predict(X), color='red', linewidth=3)

plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

```python
#Getting r square from Regression
regression.score(X, y)  #% of the variation that's explained by this model.
```
